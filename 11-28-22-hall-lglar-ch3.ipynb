## Definitions and First Examples

The text introduces Lie algebras as a more abstract notion, and I assume will show how they relate to Lie groups following this exposition. Lie algebras are usually denoted using the Fraktur characters like $\frak{g}$ and $\frak{h}$. 

> **Definition 3.1.**: A **finite-dimensional real or complex Lie algebra** is a finite-dimensional real or complex vector space $\frak{g}$, together with a map $[\cdot, \cdot]$ from $\frak{g} \times \frak{g}$ into $\frak{g}$, with the following properties: 
> 
> 1. $[\cdot, \cdot]$ is bilinear. This simply means that the function is linear in each of its inputs.
> 2. $[\cdot, \cdot]$ is skew symmetric: $[X, Y] = -[Y, X]$ for all $X, Y \in \frak{g}$. 
> 3. The **Jacobi identity** holds: For all $X, Y, Z \in \frak{g}$, 
> 
> $$[X, [Y, Z]] + [Y, [Z, X]] + [Z, [X, Y]] = 0.$$

We say that two elements $X, Y$ of a Lie algebra commute if $[X, Y] = 0$. A Lie algebra $\frak{g}$ is commutative if every pair of elements commute. 

Let $A$ be an associative algebra and let $\frak{g}$ be a subspace of $A$ such that $XY - YX \in \frak{g}$ for all $X, Y \in \frak{g}$. Then $\frak{g}$ is a Lie algebra with bracket operation given by 

$$[X, Y] = XY - YX.$$

The bilinearity and skew symmetry are easy to show for this map. The Jacobi identity can also easily be verified with a few minutes of expanding the brackets and verifying the result is zero. Apparently in a later chapter it will be shown that every Lie algebra $\frak{g}$ can be embedded into an associative algebra $A$ in such a way that the bracket becomes $XY - YX$ in what is called the Poincare-Birkhoff-Witt theorem. This is neat because the definition of the bracket is quite abstract, but this mapping allows us to work with it in a more concrete way. 

---

**Example**: Let $\frak{sl}(n;\mathbb{C})$ denote the space of all $X \in M_n(\mathbb{C})$ for which $\text{tr}(X) = 0$. Then $\frak{sl}(n;\mathbb{C})$ is a Lie algebra with bracket $[X, Y] = XY - YX$. 

By the statement above, it suffices to show that the base algebra is associative, and that $XY - YX \in \frak{g}$ for all $X, Y \in \frak{g}$ to prove this is a Lie algebra. The space of all $n \times n$ matrices is associative. To see that $XY - YX \in \frak{sl}(n;\mathbb{C})$, 

$$\text{tr}(XY - YX) = \text{tr}(XY - YX) = \text{tr}(XY) - \text{tr}(YX) = \text{tr}(XY) - \text{tr}(XY) = 0.$$

Thus we have that $\frak{sl}(n;\mathbb{C})$ is indeed a Lie algebra. 

---

Given the definition for a Lie algebra, we can now define the subalgebra too. A few other definitions that will be used later are introduced here as well. 

> **Definition 3.4**: A **subalgebra** of a real or complex Lie algebra $\frak{g}$ is a subspace $\frak{h}$ of $\frak{g}$ such that $[H_1, H_2] \in \frak{h}$ for all $H_1, H_2 \in \frak{h}$. If $\frak{g}$ is a complex Lie algebra and $\frak{h}$ is a real subspace of $\frak{g}$ which is closed under brackets, then $\frak{h}$ is said to be a **real subalgebra** of $\frak{g}$. 
> 
> A subalgebra $\frak{h}$ os a Lie algebra $\frak{g}$ is said to be an **ideal** in $\frak{g}$ if $[X, H] \in \frak{h}$ fro all $X \in \frak{g}$ and $H \in \frak{h}$. 
> 
> The **center** of a Lie algebra $\frak{g}$ is the set of all $X \in \frak{g}$ for which $[X, Y] = 0$ for all $Y \in \frak{g}$. 

> **Definition 3.5**: If $\frak{g}$ and $\frak{h}$ are Lie algebras, then a linear map $\phi: \frak{g} \rightarrow \frak{h}$ is called a **Lie algebra homomorphism** if $\phi([X, Y]) = [\phi(X), \phi(Y)]$ for all $X, Y \in \frak{g}$. If, in addition, $\phi is one-to-one and onto, then $\phi$ is called a **Lie algebra isomorphism**. A Lie algebra isomorphism of a Lie algebra with itself is called a **Lie algebra automorphism**. 

> **Definition 3.6**: If $\frak{g}$ is a Lie algebra and $X$ is an element of $\frak{g}$, define a linear map $\text{ad}_X : \frak g \rightarrow \frak g$ by
> $$\text{ad}_X(Y) = [X, Y].$$
> The map $X \mapsto \text{ad}_X$ is the **adjoint map** or **adjoint representation**. 

The author claims that this representation will be a helpful way to work through some later properties we would like to prove about Lie algebras, but for now I will skip over a few of the technical details related to the adjoint map until they are more relevant. I will just record the following proposition without proof, as it is not too hard to prove. 

> **Proposition 3.8.**: If $\frak{g}$ is a Lie algebra, then 
> $$\text{ad}_{[X, Y]} = \text{ad}_X \text{ad}_Y - \text{ad}_Y \text{ad}_X = [\text{ad}_X, \text{ad}_Y];$$
> that is, $ad: \frak g \rightarrow \text{End}(\frak g)$ is a Lie algebra homomorphism.





## Simple, Solvable, and Nilpotent Lie Algebras

In this section, special types of Lie algebras are considered and introduced. 

> **Definition 3.11.**: A Lie algebra $\frak g$ is called **irreducible** if the only ideals in $\frak g$ are $\frak g$ and $\{0\}$. A lie algebra $\frak g$ is called **simple** if it is irreducible and $\dim \frak{g} \geq 2$. 

Again, an example will help illuminate the definition a bit. 

--- 

**Example** (Proposition 3.12. in the text): The Lie algebra $\frak{sl}(2; \mathbb{C})$ is simple. 

*PROOF*: The dimension of $\frak{sl}(n;\mathbb{C})$ is 3 so it suffices to show that $\frak{sl}(2;\mathbb{C})$ is irreducible. The set containing the zero matrix $\{0\}$ is trivially an ideal, since there is no way to multiply a matrix and get something else. What remains is to show that any ideal $\frak h$ containing a non-zero element is actually equal to $\frak{sl}(n;\mathbb{C})$. 

We can express any matrix in this algebra as a linear combination of the following three basis elements (which is also why the dimension of the algebra is 3): 

$$A = \begin{bmatrix} 0 & 1 \\ 0 & 0 \end{bmatrix}, B = \begin{bmatrix} 0 & 0 \\ 1 & 0 \end{bmatrix}, C = \begin{bmatrix} 1 & 0 \\ 0 & -1 \end{bmatrix}.$$

We can verify that $[A, B] = C$, $[C, A] = 2A$, and $[C, B] = -2B$. 

Suppose now we have an ideal $\frak h$ that contains a nonzero linear combination $H$ of the above, or $H = aA + bB + cC$ for some $a, b, c \in \mathbb{C}$. To prove that $\frak h = \frak{sl}(n;\mathbb{C})$, we just need to show that each of the basis elements is in $\frak h$. 

Suppose first that $b \neq 0$. We first show that $A \in \frak h$. Since $\frak h$ is an ideal with element $H$, this implies that $[A, H] \in \frak h$. By the same argument, we also know that $[A, [A, H]] \in \frak h$. Let's see what this element is: 

$$[A, [A, H]] = [A, a[A,A] + b[A,B] + c[A,C]] = [A, bC -2A] = b[A, C] = -2bA,$$

where we have used the bilinearity of the bracket and the identities of the basis elements. From this, we can conclude that $A \in \frak h$. By the identities for the basis elements, we also know that $[B, A]$ is a nonzero multiple of $C$, and 
$$[B, [B, A]] = [B, -C] = -2B,$$ 
which implies that $B, C \in \frak h$. So we conclude that $\frak h = \frak{sl}(n;\mathbb{C})$. 

Now assume that $b = 0$ but $c \neq 0$. Then $H = aA + cC$. If we take $[A, H] = -2cA$, we see that $A \in \frak h$. By the same argument as above we can use this to show that $B, C \in \frak h$ again and conclude that $\frak h = \frak{sl}(n;\mathbb{C})$. 

Finally, assume that $b = 0$ and $c = 0$ but $a \neq 0$. Then $H = aA$. Immediately this implies that $A \in \frak h$, which we've seen shows that $\frak h = \frak{sl}(n;\mathbb{C})$ as desired.

> **Definition 3.13**: If $\frak g$ is a Lie algebra, then the **commutator ideal** in $\frak g$, denoted $[\frak g, \frak g]$ is the space of linear combinations of commutators, that is, the space of elements $Z$ in $\frak g$ that can be expressed as 
> $$Z = c_1 [X_1, Y_1] + \cdots + c_m [X_m, Y_m]$$
> for some constants $c_j$ and vectors $X_j, Y_j \in \frak g$. 

> **Definition 3.14**: For any Lie algebra $\frak g$, we define a sequence of subalgebras $\frak{g}_0, \frak{g}_1, \frak{g}_2, \ldots $ of $\frak g$ inductively as follows: $\frak{g}_0 = \frak g$, $\frak{g}_1 = [\frak{g}_0, \frak{g}_0]$, $\frak{g}_2 = [\frak{g}_1, \frak{g}_1]$, etc. These subalgebras are called the **derived series** of $\frak g$. A Lie algebra $\frak g$ is called **solvable** if $\frak{g}_j = \{0\}$ for some $j$. 

---

**Quick check**: Every $\frak{g}_j$ is an ideal in $\frak g$. 

*PROOF*: The statement can be proved inductively. Every algebra is an ideal in itself by definition, so $\frak{g}_0$ is an ideal in $\frak g$. 

For any $X, Y \in \frak g$, we have that $[X, Y] \in \frak g$ but also that $[X, Y] \in [\frak g, \frak g]$, the commutator ideal (definition 3.13). It's easy to see that $[\frak g, \frak g]$ is indeed a subalgebra of $\frak g$, and that it is also an ideal in $\frak g$, since for any $Z \in \frak g$, $[Z, [X, Y]] \in [\frak g, \frak g]$. 

More generally, suppose that for some positive integer $j$, $\frak{g}_{j-1}$ is an ideal in $\frak g$. Now define, 

$$\frak{g}_{j} = [\frak{g}_{j-1}, \frak{g}_{j-1}].$$

1. $\frak{g}_j$ is a subalgebra of $\frak g$: Take $X, Y, Z, W \in \frak{g}_{j-1}$, and consider $H = [[X, Y], [Z, W]]$. Since $\frak{g}_{j-1}$ is a subalgebra of $\frak g$, we can see that $H \in \frak{g}_j$. 

2. $\frak{g}_j$ is an ideal: Given some $G \in \frak g$, we want to verify that $[G, [X, Y]] \in \frak{g}_j$ for some $X, Y \in \frak{g}_{j-1}$. By the Jacobi identity, 

$$[G, [X, Y]] = - [X, [Y, G]] - [Y, [G, X]].$$

Since $\frak{g}_{j-1}$ is an ideal, $[G, Y], [G,X] \in \frak{g}_{j-1}$. From this, we see that the above is a linear combination of commutators in $\frak{g}_{j-1}$, implying that $\frak{g}_j$ is indeed an ideal in $\frak g$.
